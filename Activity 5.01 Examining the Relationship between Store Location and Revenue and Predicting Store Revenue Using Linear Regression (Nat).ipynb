```python
#Activity 5.02: Predicting Store Revenue Using Linear Regression

#Import libraries

import numpy as np, pandas as pd
import matplotlib.pyplot as plt, seaborn as sns
df = pd.read_csv('location_rev.csv')
df.head()
```

```python
#Create a variable, X, with the predictors

X = df[['num_competitors',\
       'median_income',\
       'num_loyalty_members',\
       'population_density',\
       'location_age']]
y = df['revenue']
```

```python
#Splot the data into a training and test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split\
(X, y, random_state = 100)
```

```python
#Create a linear regression model and fit it

from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
#Print out the model coefficient

model.coef_
```

```python
#Print out the model intercept

model.intercept_
```

```python
#Produce a prediction (3 competitiors, income 30k, 1200 loyalty, pop 2000)

single_location = pd.DataFrame({'num_competitors':[3],\
                               'median_income':[30000],\
                               'num_loyalty_members':[1200],\
                                'population_density': [2000],\
                               'location_age':[10]})

model.predict(single_location)
```

```python
#Plot the model prediction

plt.scatter(model.predict(X_test), y_test)
plt.xlabel('Model Predictions')
plt.ylabel('True Value')
plt.plot([0, 100000], [0,100000], 'k-', color = 'gray')
plt.show()
```

```python
#Calculate the correlation between the model predictions and the true values of the test data

np.corrcoef(model.predict(X_test), y_test)
```
